### 1. Scalar Field in 1D

```python
# ==============================================================================
# 단계 1: 라이브러리 임포트
# ------------------------------------------------------------------------------
# 데이터 생성을 위한 numpy와 시각화를 위한 matplotlib을 가져옵니다.
# ==============================================================================
import numpy as np
import matplotlib.pyplot as plt


# ==============================================================================
# 단계 2: 공간 정의 (좌표 그리드 생성)
# ------------------------------------------------------------------------------
# 시각화할 1차원 공간(선)의 범위와 해상도를 결정합니다.
# 예: -2π부터 2π까지 400개의 점으로 이루어진 1차원 공간을 정의합니다.
# ==============================================================================
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)


# ==============================================================================
# 단계 3: 필드(함수) 정의
# ------------------------------------------------------------------------------
# 공간의 각 지점(x)에 스칼라 값을 할당하는 함수를 정의합니다.
# 이 함수를 바꾸면 다른 1D 필드를 시각화할 수 있습니다.
# 예: 가우시안 함수: scalar_field = np.exp(-x**2)
# ==============================================================================
scalar_field = np.sin(x) * np.cos(5 * x)


# ==============================================================================
# 단계 4: 시각화 준비 (Figure 및 Axes 객체 생성)
# ------------------------------------------------------------------------------
# 그림(figure)과 그 위에 그릴 좌표축(axes)을 생성합니다.
# ==============================================================================
fig, ax = plt.subplots(figsize=(12, 6))


# ==============================================================================
# 단계 5: 필드 시각화 (데이터 플로팅)
# ------------------------------------------------------------------------------
# 1D 스칼라 필드는 y축의 높이로 표현되므로 ax.plot()을 사용합니다.
# ==============================================================================
ax.plot(x, scalar_field, label='Scalar Value')


# ==============================================================================
# 단계 6: 그래프 꾸미기 (레이블, 타이틀, 그리드 등)
# ------------------------------------------------------------------------------
# 그래프의 가독성을 높이기 위한 추가 설정입니다.
# ==============================================================================
ax.set_xlabel('X axis (Position)')
ax.set_ylabel('Scalar Value')
ax.set_title('1D Scalar Field Visualization')
ax.grid(True, linestyle='--', alpha=0.6) # 가독성을 위한 그리드 추가
ax.legend() # 범례 표시


# ==============================================================================
# 단계 7: 최종 출력
# ------------------------------------------------------------------------------
# 완성된 그래프를 화면에 표시합니다.
# ==============================================================================
plt.show()
```

### 2. Scalar Field in 2D

```python
# ==============================================================================
# 단계 1: 라이브러리 임포트
# ------------------------------------------------------------------------------
# 데이터 생성을 위한 numpy와 시각화를 위한 matplotlib을 가져옵니다.
# ==============================================================================
import numpy as np
import matplotlib.pyplot as plt


# ==============================================================================
# 단계 2: 공간 정의 (좌표 그리드 생성)
# ------------------------------------------------------------------------------
# 시각화할 2차원 공간(평면)의 범위와 해상도를 결정합니다.
# x축과 y축을 정의한 후, meshgrid를 사용해 2차원 좌표 행렬을 생성합니다.
# ==============================================================================
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)


# ==============================================================================
# 단계 3: 필드(함수) 정의
# ------------------------------------------------------------------------------
# 공간의 각 지점(x, y)에 스칼라 값을 할당하는 함수를 정의합니다.
# 이 함수를 바꾸면 다른 2D 필드를 시각화할 수 있습니다.
# 예: 파동 함수 f(x,y) = sin(sqrt(x^2 + y^2))
# ==============================================================================
# 원점으로부터의 거리 R을 계산하여 동심원 형태의 필드를 만듭니다.
R = np.sqrt(X**2 + Y**2)
scalar_field = np.sin(R)


# ==============================================================================
# 단계 4: 시각화 준비 (Figure 및 Axes 객체 생성)
# ------------------------------------------------------------------------------
# 그림(figure)과 그 위에 그릴 좌표축(axes)을 생성합니다.
# ==============================================================================
fig, ax = plt.subplots(figsize=(10, 8))


# ==============================================================================
# 단계 5: 필드 시각화 (데이터 플로팅)
# ------------------------------------------------------------------------------
# 2D 스칼라 필드는 색상으로 표현하므로 ax.pcolormesh() 또는 ax.contourf()를 사용합니다.
# cmap은 컬러맵을 지정합니다. 'viridis', 'plasma', 'jet', 'coolwarm' 등 다양합니다.
# ==============================================================================
mesh = ax.pcolormesh(X, Y, scalar_field, cmap='viridis', shading='auto')


# ==============================================================================
# 단계 6: 그래프 꾸미기 (레이블, 타이틀, 컬러바 등)
# ------------------------------------------------------------------------------
# 그래프의 가독성을 높이기 위한 추가 설정입니다.
# ==============================================================================
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('2D Scalar Field Visualization')
ax.set_aspect('equal', adjustable='box') # x, y축 비율을 동일하게 설정

# 컬러바는 스칼라 값과 색상을 매핑해주는 필수적인 요소입니다.
fig.colorbar(mesh, ax=ax, label='Scalar Value')


# ==============================================================================
# 단계 7: 최종 출력
# ------------------------------------------------------------------------------
# 완성된 그래프를 화면에 표시합니다.
# ==============================================================================
plt.show()
```

### 3. Vector Field in 2D

```python
# ==============================================================================
# 단계 1: 라이브러리 임포트
# ------------------------------------------------------------------------------
# 데이터 생성을 위한 numpy와 시각화를 위한 matplotlib을 가져옵니다.
# ==============================================================================
import numpy as np
import matplotlib.pyplot as plt


# ==============================================================================
# 단계 2: 공간 정의 (좌표 그리드 생성)
# ------------------------------------------------------------------------------
# 시각화할 2차원 공간(평면)의 범위와 해상도를 결정합니다.
# 벡터 필드는 화살표가 많으면 복잡해 보이므로, 해상도(점의 개수)를 스칼라 필드보다
# 낮게 설정하는 것이 가독성에 좋습니다.
# ==============================================================================
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)


# ==============================================================================
# 단계 3: 필드(함수) 정의
# ------------------------------------------------------------------------------
# 공간의 각 지점(x, y)에 벡터 값을 할당합니다.
# 벡터는 x방향 성분 'U'와 y방향 성분 'V'로 구성됩니다.
#
# 이 함수들을 바꾸면 다른 2D 벡터 필드를 시각화할 수 있습니다.
# 예: 중심으로 향하는 필드(Sink): U = -X, V = -Y
# 예: 소용돌이와 발산을 합친 필드: U = X - Y, V = X + Y
# ==============================================================================
# 여기서는 반시계 방향으로 회전하는 벡터 필드 F = (-y, x)를 정의합니다.
U = -Y  # x방향 성분
V = X   # y방향 성분


# ==============================================================================
# 단계 4: 시각화 준비 (Figure 및 Axes 객체 생성)
# ------------------------------------------------------------------------------
# 그림(figure)과 그 위에 그릴 좌표축(axes)을 생성합니다.
# ==============================================================================
fig, ax = plt.subplots(figsize=(10, 10))


# ==============================================================================
# 단계 5: 필드 시각화 (데이터 플로팅)
# ------------------------------------------------------------------------------
# 2D 벡터 필드는 각 (X, Y) 지점에 (U, V) 벡터를 화살표로 그리는 ax.quiver()를 사용합니다.
#
# [다른 시각화 아이디어]
# 유선(Streamline)을 그리고 싶다면 ax.streamplot()을 사용할 수 있습니다.
#   ax.streamplot(X, Y, U, V, color='blue')
# ==============================================================================
# X, Y는 화살표의 시작 위치, U, V는 화살표의 방향과 크기를 결정합니다.
ax.quiver(X, Y, U, V)


# ==============================================================================
# 단계 6: 그래프 꾸미기 (레이블, 타이틀 등)
# ------------------------------------------------------------------------------
# 그래프의 가독성을 높이기 위한 추가 설정입니다.
# ==============================================================================
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('2D Vector Field Visualization')

# 벡터의 방향이 왜곡되지 않도록 x, y축의 비율을 동일하게 설정하는 것이 매우 중요합니다.
ax.set_aspect('equal', adjustable='box')


# ==============================================================================
# 단계 7: 최종 출력
# ------------------------------------------------------------------------------
# 완성된 그래프를 화면에 표시합니다.
# ==============================================================================
plt.show()
```

### 4. Scalar Field in 3D

### 5. Vector Field in 3D

```python
# ==============================================================================
# 단계 1: 라이브러리 임포트
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
# 3D 축을 사용하기 위해 Axes3D를 임포트합니다 (직접 사용하지 않아도 등록됨).
from mpl_toolkits.mplot3d import Axes3D

# ==============================================================================
# 단계 2: 공간 정의 (좌표 그리드 생성)
# ------------------------------------------------------------------------------
# 전기장을 계산할 3차원 공간의 좌표 격자를 생성합니다.
# ==============================================================================
x, y, z = np.meshgrid(np.arange(-1, 1.2, 0.4),
                      np.arange(-1, 1.2, 0.4),
                      np.arange(-1, 1.2, 0.4))

# ==============================================================================
# 단계 3: 필드(함수) 정의
# ------------------------------------------------------------------------------
# 3차원 공간의 각 지점 (x, y, z)에 벡터 값을 할당하는 함수를 정의합니다.
# 이 예제에서는 원점에 있는 점전하의 전기장 벡터(Ex, Ey, Ez)를 계산합니다.
# ==============================================================================
# 점전하의 위치 및 전하량 설정
charge_pos = np.array([0, 0, 0])
charge = 1.0

# 각 지점에서 전하까지의 거리 벡터 및 거리의 세제곱 계산
rx = x - charge_pos[0]
ry = y - charge_pos[1]
rz = z - charge_pos[2]
r_cubed = (rx**2 + ry**2 + rz**2)**1.5

# 전하 위치에서 0으로 나누어지는 것을 방지하기 위한 처리
r_cubed[r_cubed == 0] = 1e-6

# 각 지점에서의 전기장 벡터(Ex, Ey, Ez) 계산
Ex = charge * rx / r_cubed
Ey = charge * ry / r_cubed
Ez = charge * rz / r_cubed

# ==============================================================================
# 단계 4: 시각화 준비 (Figure 및 3D Axes 객체 생성)
# ------------------------------------------------------------------------------
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# ==============================================================================
# 단계 5: 필드 시각화 (데이터 플로팅)
# ------------------------------------------------------------------------------
# 3D 벡터 필드는 quiver 함수를 사용하여 화살표로 표현합니다.
# x, y, z: 화살표의 시작점
# Ex, Ey, Ez: 각 화살표의 방향과 크기를 나타내는 벡터 성분
# length: 화살표의 전체적인 길이 조절
# normalize: 모든 화살표의 길이를 같게 정규화하여 방향을 명확히 함
# ==============================================================================
ax.quiver(x, y, z, Ex, Ey, Ez, length=0.2, normalize=True, color='blue')

# 시각적 구분을 위해 원점에 있는 점전하를 빨간 점으로 표시합니다.
ax.scatter([0], [0], [0], color='red', s=100, label='Point Charge')

# ==============================================================================
# 단계 6: 그래프 꾸미기 (레이블, 타이틀, 범례 등)
# ------------------------------------------------------------------------------
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Electric Field of a Point Charge')
ax.legend()

# ==============================================================================
# 단계 7: 최종 출력
# ------------------------------------------------------------------------------
plt.show()
```

```python
# ==============================================================================
# 단계 1: 라이브러리 임포트
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ==============================================================================
# 단계 2: 공간 정의 (좌표 그리드 생성)
# ------------------------------------------------------------------------------
# 화살표가 너무 빽빽하지 않도록 해상도를 매우 낮게 설정합니다.
# ==============================================================================
x = np.linspace(-5, 5, 8)
y = np.linspace(-5, 5, 8)
z = np.linspace(-5, 5, 8)
X, Y, Z = np.meshgrid(x, y, z)

# ==============================================================================
# 단계 3: 필드(함수) 정의
# ------------------------------------------------------------------------------
# 3차원 벡터 필드는 x, y, z 세 방향의 성분 (U, V, W)으로 구성됩니다.
# 예: z축을 중심으로 회전하며 아래로 수렴하는 필드 (토네이도 모양)
# ==============================================================================
U = -Y
V = X
W = -2 * np.ones_like(Z) # 모든 지점에서 z방향으로 -2의 힘

# ==============================================================================
# 단계 4: 시각화 준비 (Figure 및 3D Axes 객체 생성)
# ------------------------------------------------------------------------------
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# ==============================================================================
# 단계 5: 필드 시각화 (데이터 플로팅)
# ------------------------------------------------------------------------------
# 3D 벡터 필드는 3D quiver() 함수를 사용합니다.
# length: 화살표의 전체적인 길이 스케일
# normalize: 모든 화살표 길이를 같게 만들어 방향을 보기 좋게 할지 여부
# ==============================================================================
ax.quiver(X, Y, Z, U, V, W, length=0.5, normalize=True)

# ==============================================================================
# 단계 6: 그래프 꾸미기 (레이블, 타이틀 등)
# ------------------------------------------------------------------------------
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Vector Field Visualization')

# ==============================================================================
# 단계 7: 최종 출력
# ------------------------------------------------------------------------------
plt.show()
```

### 6. For simple Visualization

```python
# ==============================================================================
# 단계 1: 라이브러리 임포트
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ==============================================================================
# 단계 2: 공간 정의 (좌표 그리드 생성)
# ------------------------------------------------------------------------------
# 렌더링 부하를 줄이기 위해 해상도를 적절하게 설정합니다.
# ==============================================================================
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
z = np.linspace(-5, 5, 30)
X, Y, Z = np.meshgrid(x, y, z)

# ==============================================================================
# 단계 3: 필드(함수) 정의
# ------------------------------------------------------------------------------
# 3차원 가우시안 분포 함수를 정의합니다.
# ==============================================================================
scalar_field = np.exp(-(X**2 + Y**2 + Z**2) / 5)

# ==============================================================================
# 단계 4: 시각화 준비 (Figure 및 3D Axes 객체 생성)
# ------------------------------------------------------------------------------
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# ==============================================================================
# 단계 5: 필드 시각화 (데이터 플로팅) - << 개선된 부분
# ------------------------------------------------------------------------------
# 스칼라 값이 특정 임계값(threshold) 이상인 데이터만 필터링하여 선택합니다.
# 이 작업을 통해 바깥쪽의 희미한 '안개'를 제거하고 핵심만 볼 수 있습니다.
# ==============================================================================
threshold = 0.3  # 이 값을 조절하여 표시할 데이터의 양을 결정 (0.1 ~ 0.9)
mask = scalar_field > threshold

# 필터링된 데이터만 사용하여 산점도를 그립니다.
scatter_plot = ax.scatter(
    X[mask], Y[mask], Z[mask],  # 마스크를 적용하여 조건에 맞는 좌표만 선택
    c=scalar_field[mask],      # 해당하는 스칼라 값만 색상으로 사용
    cmap='viridis',
    alpha=0.7                  # 필터링 했으므로 투명도를 약간 높여 선명하게 만듦
)

# ==============================================================================
# 단계 6: 그래프 꾸미기 (레이블, 타이틀, 컬러바 등)
# ------------------------------------------------------------------------------
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Improved 3D Scalar Field (Thresholded)') # << 제목 변경
fig.colorbar(scatter_plot, shrink=0.6, label='Scalar Value')

# ==============================================================================
# 단계 7: 최종 출력
# ------------------------------------------------------------------------------
plt.show()
```

```python
# ==============================================================================
# 단계 1: 라이브러리 임포트
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ==============================================================================
# 단계 2: 공간 정의 (좌표 그리드 생성)
# ------------------------------------------------------------------------------
# 화살표가 너무 빽빽하지 않도록 해상도를 매우 낮게 설정합니다.
# ==============================================================================
x = np.linspace(-5, 5, 8)
y = np.linspace(-5, 5, 8)
z = np.linspace(-5, 5, 8)
X, Y, Z = np.meshgrid(x, y, z)

# ==============================================================================
# 단계 3: 필드(함수) 정의
# ------------------------------------------------------------------------------
# 3차원 벡터 필드는 x, y, z 세 방향의 성분 (U, V, W)으로 구성됩니다.
# 예: z축을 중심으로 회전하며 아래로 수렴하는 필드 (토네이도 모양)
# ==============================================================================
U = -Y
V = X
W = -2 * np.ones_like(Z) # 모든 지점에서 z방향으로 -2의 힘

# ==============================================================================
# 단계 4 (수정): 1x3 형태의 서브플롯(subplot) 생성
# ------------------------------------------------------------------------------
# 3개의 3D 그래프를 가로로 나란히 배치합니다.
# ==============================================================================
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 7), subplot_kw={'projection': '3d'})
fig.suptitle('Truly Improved 3D Vector Field (Separated Slices)', fontsize=16)

# ==============================================================================
# 단계 5 (수정): 각 서브플롯에 하나의 단면만 그리기
# ------------------------------------------------------------------------------
# 중앙 인덱스를 계산합니다.
x_slice_idx = len(x) // 2
y_slice_idx = len(y) // 2
z_slice_idx = len(z) // 2

# 플롯 1: XY 평면 (z=0)
ax1.quiver(X[:, :, z_slice_idx], Y[:, :, z_slice_idx], Z[:, :, z_slice_idx],
           U[:, :, z_slice_idx], V[:, :, z_slice_idx], W[:, :, z_slice_idx],
           length=0.8, normalize=True, color='red')
ax1.set_title('Slice: XY Plane (z=0)')

# 플롯 2: XZ 평면 (y=0)
ax2.quiver(X[:, y_slice_idx, :], Y[:, y_slice_idx, :], Z[:, y_slice_idx, :],
           U[:, y_slice_idx, :], V[:, y_slice_idx, :], W[:, y_slice_idx, :],
           length=0.8, normalize=True, color='blue')
ax2.set_title('Slice: XZ Plane (y=0)')

# 플롯 3: YZ 평면 (x=0)
ax3.quiver(X[x_slice_idx, :, :], Y[x_slice_idx, :, :], Z[x_slice_idx, :, :],
           U[x_slice_idx, :, :], V[x_slice_idx, :, :], W[x_slice_idx, :, :],
           length=0.8, normalize=True, color='green')
ax3.set_title('Slice: YZ Plane (x=0)')

# ==============================================================================
# 단계 6: 각 그래프 꾸미기
# ------------------------------------------------------------------------------
for ax in [ax1, ax2, ax3]:
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_xlim([-5, 5]); ax.set_ylim([-5, 5]); ax.set_zlim([-5, 5])

# ==============================================================================
# 단계 7: 최종 출력
# ------------------------------------------------------------------------------
plt.tight_layout(rect=[0, 0, 1, 0.95]) # suptitle과 겹치지 않게 조정
plt.show()
```
