### 1. Scalar Field in 1D

```python
# ==============================================================================
#  1:  
# ------------------------------------------------------------------------------
#    numpy   matplotlib .
# ==============================================================================
import numpy as np
import matplotlib.pyplot as plt


# ==============================================================================
#  2:   (  )
# ------------------------------------------------------------------------------
#  1 ()   .
# : -2 2 400   1  .
# ==============================================================================
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)


# ==============================================================================
#  3: () 
# ------------------------------------------------------------------------------
#   (x)     .
#     1D    .
# :  : scalar_field = np.exp(-x**2)
# ==============================================================================
scalar_field = np.sin(x) * np.cos(5 * x)


# ==============================================================================
#  4:   (Figure  Axes  )
# ------------------------------------------------------------------------------
# (figure)    (axes) .
# ==============================================================================
fig, ax = plt.subplots(figsize=(12, 6))


# ==============================================================================
#  5:   ( )
# ------------------------------------------------------------------------------
# 1D   y   ax.plot() .
# ==============================================================================
ax.plot(x, scalar_field, label='Scalar Value')


# ==============================================================================
#  6:   (, ,  )
# ------------------------------------------------------------------------------
#      .
# ==============================================================================
ax.set_xlabel('X axis (Position)')
ax.set_ylabel('Scalar Value')
ax.set_title('1D Scalar Field Visualization')
ax.grid(True, linestyle='--', alpha=0.6) #    
ax.legend() #  


# ==============================================================================
#  7:  
# ------------------------------------------------------------------------------
#    .
# ==============================================================================
plt.show()
```

### 2. Scalar Field in 2D

```python
# ==============================================================================
#  1:  
# ------------------------------------------------------------------------------
#    numpy   matplotlib .
# ==============================================================================
import numpy as np
import matplotlib.pyplot as plt


# ==============================================================================
#  2:   (  )
# ------------------------------------------------------------------------------
#  2 ()   .
# x y  , meshgrid  2   .
# ==============================================================================
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)


# ==============================================================================
#  3: () 
# ------------------------------------------------------------------------------
#   (x, y)     .
#     2D    .
# :   f(x,y) = sin(sqrt(x^2 + y^2))
# ==============================================================================
#   R     .
R = np.sqrt(X**2 + Y**2)
scalar_field = np.sin(R)


# ==============================================================================
#  4:   (Figure  Axes  )
# ------------------------------------------------------------------------------
# (figure)    (axes) .
# ==============================================================================
fig, ax = plt.subplots(figsize=(10, 8))


# ==============================================================================
#  5:   ( )
# ------------------------------------------------------------------------------
# 2D     ax.pcolormesh()  ax.contourf() .
# cmap  . 'viridis', 'plasma', 'jet', 'coolwarm'  .
# ==============================================================================
mesh = ax.pcolormesh(X, Y, scalar_field, cmap='viridis', shading='auto')


# ==============================================================================
#  6:   (, ,  )
# ------------------------------------------------------------------------------
#      .
# ==============================================================================
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('2D Scalar Field Visualization')
ax.set_aspect('equal', adjustable='box') # x, y   

#       .
fig.colorbar(mesh, ax=ax, label='Scalar Value')


# ==============================================================================
#  7:  
# ------------------------------------------------------------------------------
#    .
# ==============================================================================
plt.show()
```

### 3. Vector Field in 2D

```python
# ==============================================================================
#  1:  
# ------------------------------------------------------------------------------
#    numpy   matplotlib .
# ==============================================================================
import numpy as np
import matplotlib.pyplot as plt


# ==============================================================================
#  2:   (  )
# ------------------------------------------------------------------------------
#  2 ()   .
#      , ( )  
#     .
# ==============================================================================
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)


# ==============================================================================
#  3: () 
# ------------------------------------------------------------------------------
#   (x, y)   .
#  x  'U' y  'V' .
#
#     2D     .
# :   (Sink): U = -X, V = -Y
# :    : U = X - Y, V = X + Y
# ==============================================================================
#       F = (-y, x) .
U = -Y  # x 
V = X   # y 


# ==============================================================================
#  4:   (Figure  Axes  )
# ------------------------------------------------------------------------------
# (figure)    (axes) .
# ==============================================================================
fig, ax = plt.subplots(figsize=(10, 10))


# ==============================================================================
#  5:   ( )
# ------------------------------------------------------------------------------
# 2D    (X, Y)  (U, V)    ax.quiver() .
#
# [  ]
# (Streamline)   ax.streamplot()   .
#   ax.streamplot(X, Y, U, V, color='blue')
# ==============================================================================
# X, Y   , U, V    .
ax.quiver(X, Y, U, V)


# ==============================================================================
#  6:   (,  )
# ------------------------------------------------------------------------------
#      .
# ==============================================================================
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('2D Vector Field Visualization')

#     x, y      .
ax.set_aspect('equal', adjustable='box')


# ==============================================================================
#  7:  
# ------------------------------------------------------------------------------
#    .
# ==============================================================================
plt.show()
```

### 4. Scalar Field in 3D

```python
# ==============================================================================
#  1:  
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
# 3D    Axes3D  (   ).
from mpl_toolkits.mplot3d import Axes3D

# ==============================================================================
#  2:   (  )
# ------------------------------------------------------------------------------
# 3    z , 3D meshgrid .
#       .
# ==============================================================================
x = np.linspace(-5, 5, 25)
y = np.linspace(-5, 5, 25)
z = np.linspace(-5, 5, 25)
X, Y, Z = np.meshgrid(x, y, z)

# ==============================================================================
#  3: () 
# ------------------------------------------------------------------------------
# 3    (x, y, z)     .
# : 3D  
# ==============================================================================
scalar_field = np.exp(-(X**2 + Y**2 + Z**2) / 5)

# ==============================================================================
#  4:   (Figure  3D Axes  )
# ------------------------------------------------------------------------------
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# ==============================================================================
#  5:   ( )
# ------------------------------------------------------------------------------
# 3D   3D (scatter)   .
# c:      
# cmap:  
# alpha:   (    )
# ==============================================================================
scatter_plot = ax.scatter(X, Y, Z, c=scalar_field.flatten(), cmap='viridis', alpha=0.2)

# ==============================================================================
#  6:   (, ,  )
# ------------------------------------------------------------------------------
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Scalar Field Visualization')
fig.colorbar(scatter_plot, shrink=0.6, label='Scalar Value')

# ==============================================================================
#  7:  
# ------------------------------------------------------------------------------
plt.show()
```

```python
# ==============================================================================
#  1:  
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
# 3D    Axes3D  (   ).
from mpl_toolkits.mplot3d import Axes3D

# ==============================================================================
#  2:   (  )
# ------------------------------------------------------------------------------
# 3    x, y, z   ,
# np.meshgrid  3D   .
# ==============================================================================
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
z = np.linspace(-5, 5, 20)
X, Y, Z = np.meshgrid(x, y, z)

# ==============================================================================
#  3: () 
# ------------------------------------------------------------------------------
# 3    (X, Y, Z)     .
#   sin(X) * cos(Y) * sin(Z)  .
# ==============================================================================
scalar_field = np.sin(X) * np.cos(Y) * np.sin(Z)

# ==============================================================================
#  4:   (Figure  3D Axes  )
# ------------------------------------------------------------------------------
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# ==============================================================================
#  5:   ( )
# ------------------------------------------------------------------------------
# 3D   3D (scatter)   .
# c:       (flatten() 1  )
# cmap:   (viridis)
# alpha:   (    )
# ==============================================================================
scatter = ax.scatter(X, Y, Z, c=scalar_field.flatten(), cmap='viridis', alpha=0.5)

# ==============================================================================
#  6:   (, ,  )
# ------------------------------------------------------------------------------
#       .
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Scalar Field Visualization')

#           .
fig.colorbar(scatter, shrink=0.5, aspect=5, label='Scalar Value')

# ==============================================================================
#  7:  
# ------------------------------------------------------------------------------
plt.show()
```

### 5. Vector Field in 3D

```python
# ==============================================================================
#  1:  
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
# 3D    Axes3D  (   ).
from mpl_toolkits.mplot3d import Axes3D

# ==============================================================================
#  2:   (  )
# ------------------------------------------------------------------------------
#   3    .
# ==============================================================================
x, y, z = np.meshgrid(np.arange(-1, 1.2, 0.4),
                      np.arange(-1, 1.2, 0.4),
                      np.arange(-1, 1.2, 0.4))

# ==============================================================================
#  3: () 
# ------------------------------------------------------------------------------
# 3    (x, y, z)     .
#       (Ex, Ey, Ez) .
# ==============================================================================
#     
charge_pos = np.array([0, 0, 0])
charge = 1.0

#         
rx = x - charge_pos[0]
ry = y - charge_pos[1]
rz = z - charge_pos[2]
r_cubed = (rx**2 + ry**2 + rz**2)**1.5

#   0     
r_cubed[r_cubed == 0] = 1e-6

#    (Ex, Ey, Ez) 
Ex = charge * rx / r_cubed
Ey = charge * ry / r_cubed
Ez = charge * rz / r_cubed

# ==============================================================================
#  4:   (Figure  3D Axes  )
# ------------------------------------------------------------------------------
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# ==============================================================================
#  5:   ( )
# ------------------------------------------------------------------------------
# 3D   quiver    .
# x, y, z:  
# Ex, Ey, Ez:       
# length:    
# normalize:        
# ==============================================================================
ax.quiver(x, y, z, Ex, Ey, Ez, length=0.2, normalize=True, color='blue')

#         .
ax.scatter([0], [0], [0], color='red', s=100, label='Point Charge')

# ==============================================================================
#  6:   (, ,  )
# ------------------------------------------------------------------------------
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Electric Field of a Point Charge')
ax.legend()

# ==============================================================================
#  7:  
# ------------------------------------------------------------------------------
plt.show()
```

### 6. For simple Visualization

```python
# ==============================================================================
#  1:  
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ==============================================================================
#  2:   (  )
# ------------------------------------------------------------------------------
#       .
# ==============================================================================
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
z = np.linspace(-5, 5, 30)
X, Y, Z = np.meshgrid(x, y, z)

# ==============================================================================
#  3: () 
# ------------------------------------------------------------------------------
# 3    .
# ==============================================================================
scalar_field = np.exp(-(X**2 + Y**2 + Z**2) / 5)

# ==============================================================================
#  4:   (Figure  3D Axes  )
# ------------------------------------------------------------------------------
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# ==============================================================================
#  5:   ( ) - <<  
# ------------------------------------------------------------------------------
#    (threshold)    .
#      ''     .
# ==============================================================================
threshold = 0.3  #        (0.1 ~ 0.9)
mask = scalar_field > threshold

#     .
scatter_plot = ax.scatter(
    X[mask], Y[mask], Z[mask],  #      
    c=scalar_field[mask],      #     
    cmap='viridis',
    alpha=0.7                  #       
)

# ==============================================================================
#  6:   (, ,  )
# ------------------------------------------------------------------------------
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Improved 3D Scalar Field (Thresholded)') # <<  
fig.colorbar(scatter_plot, shrink=0.6, label='Scalar Value')

# ==============================================================================
#  7:  
# ------------------------------------------------------------------------------
plt.show()
```

```python
# ==============================================================================
#  1:  
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ==============================================================================
#  2:   (  )
# ------------------------------------------------------------------------------
#        .
# ==============================================================================
x = np.linspace(-5, 5, 8)
y = np.linspace(-5, 5, 8)
z = np.linspace(-5, 5, 8)
X, Y, Z = np.meshgrid(x, y, z)

# ==============================================================================
#  3: () 
# ------------------------------------------------------------------------------
# 3   x, y, z    (U, V, W) .
# : z      ( )
# ==============================================================================
U = -Y
V = X
W = -2 * np.ones_like(Z) #   z -2 

# ==============================================================================
#  4 (): 1x3  (subplot) 
# ------------------------------------------------------------------------------
# 3 3D    .
# ==============================================================================
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 7), subplot_kw={'projection': '3d'})
fig.suptitle('Truly Improved 3D Vector Field (Separated Slices)', fontsize=16)

# ==============================================================================
#  5 ():     
# ------------------------------------------------------------------------------
#   .
x_slice_idx = len(x) // 2
y_slice_idx = len(y) // 2
z_slice_idx = len(z) // 2

#  1: XY  (z=0)
ax1.quiver(X[:, :, z_slice_idx], Y[:, :, z_slice_idx], Z[:, :, z_slice_idx],
           U[:, :, z_slice_idx], V[:, :, z_slice_idx], W[:, :, z_slice_idx],
           length=0.8, normalize=True, color='red')
ax1.set_title('Slice: XY Plane (z=0)')

#  2: XZ  (y=0)
ax2.quiver(X[:, y_slice_idx, :], Y[:, y_slice_idx, :], Z[:, y_slice_idx, :],
           U[:, y_slice_idx, :], V[:, y_slice_idx, :], W[:, y_slice_idx, :],
           length=0.8, normalize=True, color='blue')
ax2.set_title('Slice: XZ Plane (y=0)')

#  3: YZ  (x=0)
ax3.quiver(X[x_slice_idx, :, :], Y[x_slice_idx, :, :], Z[x_slice_idx, :, :],
           U[x_slice_idx, :, :], V[x_slice_idx, :, :], W[x_slice_idx, :, :],
           length=0.8, normalize=True, color='green')
ax3.set_title('Slice: YZ Plane (x=0)')

# ==============================================================================
#  6:   
# ------------------------------------------------------------------------------
for ax in [ax1, ax2, ax3]:
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_xlim([-5, 5]); ax.set_ylim([-5, 5]); ax.set_zlim([-5, 5])

# ==============================================================================
#  7:  
# ------------------------------------------------------------------------------
plt.tight_layout(rect=[0, 0, 1, 0.95]) # suptitle   
plt.show()
```
