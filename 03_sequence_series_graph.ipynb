# Level Set

### iso-contour (Îì±ÏúÑÍ≥°ÏÑ†) : {(x,y)|f(x,y) = h; hÎäî ÏÉÅÏàò. ÏïÑÎûòÎèÑ ÎßàÏ∞¨Í∞ÄÏßÄ}
### iso-surface (Îì±ÏúÑÍ≥°Î©¥) : {(x,y)|f(x,y,z) = h}
### level set : {f=h} = {x in D | f(x) = h; Ïó¨Í∏∞ÏÑú xÎäî Îã®Î≥ÄÏàòÍ∞Ä ÏïÑÎãàÎùº Î≤°ÌÑ∞ in R^n}

```python
#Îã§Î≥ÄÏàò Ìï®Ïàò Í∑∏Î¶¨Í∏∞

#1. Í≥°Î©¥ Í∑∏ÎåÄÎ°ú Í∑∏Î¶¨Îäî Î≤ï

## f(x,y) = sin(sqrt(x^2+y^2))Ïùò Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
import numpy as np
import  matplotlib.pyplot as plt

x = np.linspace(-5, 5)
y = np.linspace(-5, 5)
xx, yy = np.meshgrid(x, y)
zz = np.sin(np.sqrt(xx**2+yy**2))

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(xx, yy, zz, cmap='jet')
plt.show()
```

```python
# np.meshgrid(x, y)Îäî Îëê Í∞úÏùò 1Ï∞®Ïõê Î∞∞Ïó¥ xÏôÄ yÎ•º Î∞õÏïÑÏÑú, 2Ï∞®Ïõê Ï¢åÌëú ÌèâÎ©¥ÏùÑ Íµ¨ÏÑ±ÌïòÎäî Í≤©Ïûê(grid)Î•º ÎßåÎì§Ïñ¥Ï£ºÎäî Ìï®Ïàò.
# Í∞ÅÍ∞Å (x,y) Í≥±ÏßëÌï©Ïùò Î∞∞Ïó¥ÏóêÏÑú xÏ¢åÌëú(xx)ÏôÄ yÏ¢åÌëú(yy)Îßå ÎÇòÌÉÄÎÇ∏ Í≤É

x = np.array([1, 2, 3])
y = np.array([10, 20])
xx, yy = np.meshgrid(x, y)

print(xx) #Ïù¥Ï∞®Ïõê Î∞∞Ïó¥Ïùò xÏ¢åÌëúÎßå Î™®ÏùÄ Í≤É
print()
print(yy) #Ïù¥Ï∞®Ïõê Î∞∞Ïó¥Ïùò yÏ¢åÌëúÎßå Î™®ÏùÄ Í≤É
print()
```

üîç Ïôú Ïù¥Îü∞ Íµ¨Ï°∞Í∞Ä ÎêòÎäîÍ∞Ä?

xÎäî 3Í∞ú ‚Üí xÏ∂ïÏóê 3Í∞úÏùò Ï†ê

yÎäî 2Í∞ú ‚Üí yÏ∂ïÏóê 2Í∞úÏùò Ï†ê

Í∑∏ÎûòÏÑú Ï†ÑÏ≤¥ Ï¢åÌëúÎäî Ï¥ù 2Ìñâ √ó 3Ïó¥ = 6Í∞ú Ï†êÏù¥ ÏÉùÍ≤®Ïïº Ìï¥.

y\x	  1	       2	  3
10	(1,10)	(2,10)	(3,10)
20	(1,20)	(2,20)	(3,20)
Ïù¥Í±∏ ÌëúÌòÑÌïòÎ†§Î©¥:

xx: Í∞Å ÏúÑÏπòÏùò xÍ∞í ‚Üí Í∞Å ÌñâÏù¥ xÏùò Î≥µÏÇ¨Î≥∏

yy: Í∞Å ÏúÑÏπòÏùò yÍ∞í ‚Üí Í∞Å Ïó¥Ïù¥ yÏùò Î≥µÏÇ¨Î≥∏

Ï¶â, xx[i][j]ÏôÄ yy[i][j]Îäî (x_j, y_i)Î•º ÎÇòÌÉÄÎÇ¥Îäî Í±∞Ïïº.

```python
#Îã§Î≥ÄÏàò Ìï®Ïàò Í∑∏Î¶¨Í∏∞

#2. Level SetÏùÑ ÌôúÏö©ÌïòÎäî Î≤ï

import numpy as np
import matplotlib.pyplot as plt

# ÏúÑÏôÄ ÎèôÏùºÌïòÍ≤å Ïù¥Î≥ÄÏàò Ìï®Ïàò ÎßåÎìúÎäî ÏΩîÎìú. Í∑∏ÎûòÌîÑÎ°ú ÏãúÍ∞ÅÌôîÎßå Ïïà ÌñàÏùÑ Îøê
x = np.linspace(-5, 5)
y = np.linspace(-5, 5)
xx, yy = np.meshgrid(x,y)
zz = np.sin(np.sqrt(xx**2+yy**2))

fig = plt.figure()
ax = plt.axes() # Ïó¨Í∏∞ÏÑú Î∂ÄÌÑ∞ Îã¨ÎùºÏßÑÎã§.
ax.contour(xx, yy, zz, 10, cmap='jet') #ÌïµÏã¨Ï∞®Ïù¥. 4Î≤àÏû¨ Ïù∏ÏàòÏù∏ 10ÏùÄ Î†àÎ≤®(Îì±Í≥†ÏÑ†) Í∞úÏàò. cmapÏùÄ ÏÉâÏÉÅ
ax.set_aspect('equal') #Í∞ÄÎ°úÏ∂ï ÏÑ∏Î°úÏ∂ï ÎπÑÏú® Í∞ôÍ≤å
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes() 
ax.contour(xx, yy, zz, [-0.4, -0.2, 0.0, 0.2, 0.4], cmap='jet') #4Î≤àÏß∏ Ïù∏ÏàòÏóê Î¶¨Ïä§Ìä∏ÎÇ¥ 5Í∞úÏùò ÏõêÏÜå(f=hÏó≠Ìï†Ïùò h 5Í∞ú)Î°ú Î†àÎ≤®ÏÖã Î∂ÄÏó¨
ax.set_aspect('equal') 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ÏòàÏãú 3Î≥ÄÏàò Ìï®Ïàò: f(x, y, z) = x**2 + y**2 + z**2
def f(x, y, z):
    return x**2 + y**2 + z**2

x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
xx, yy = np.meshgrid(x, y)

# z Í∞íÏùÑ Í≥†Ï†ïÌïòÎ©∞ Ïó¨Îü¨ Í∞úÏùò Îì±Í≥†ÏÑ† ÌîåÎ°Ø ÏÉùÏÑ±
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

z_values = [0, 2, 4] # ÌôïÏù∏Ìï† z Í∞íÎì§

for i, z_val in enumerate(z_values):
    zz = f(xx, yy, z_val)
    ax = axes[i]
    contour = ax.contour(xx, yy, zz, cmap='jet')
    ax.set_title(f'Contour plot at z = {z_val}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_aspect('equal')
    fig.colorbar(contour, ax=ax)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs # Í∞ÄÏÉÅ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±ÏùÑ ÏúÑÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨

# 1. Í∞ÄÏÉÅ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
# 4Í∞úÏùò ÌäπÏÑ±(Î≥ÄÏàò)ÏùÑ Í∞ÄÏßÑ 3Í∞úÏùò ÌÅ¥Îü¨Ïä§ÌÑ∞(Í∑∏Î£π) Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
X, y = make_blobs(n_samples=300, n_features=4, centers=3, cluster_std=2.0, random_state=42)

# XÎäî (300, 4) ÌÅ¨Í∏∞Ïùò Î∞∞Ïó¥ (ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞)
# yÎäî (300,) ÌÅ¨Í∏∞Ïùò Î∞∞Ïó¥ (0, 1, 2 Ï§ë ÌïòÎÇòÏùò Í∞íÏùÑ Í∞ÄÏßÄÎäî Í∑∏Î£π Ï†ïÎ≥¥)

# 2. Îç∞Ïù¥ÌÑ∞ Ï†ïÍ∑úÌôî
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. PCA Ï†ÅÏö© (4Ï∞®Ïõê -> 2Ï∞®Ïõê)
# n_components=2 : Îç∞Ïù¥ÌÑ∞Î•º 2Í∞úÏùò Ï£ºÏÑ±Î∂ÑÏúºÎ°ú Ï∂ïÏÜå
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled) # PCAÎ•º Ï†ÅÏö©ÌïòÏó¨ Î≥ÄÌôò

# Ï£ºÏÑ±Î∂ÑÏúºÎ°ú Î≥ÄÌôòÎêú Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏
print("ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ ÌòïÌÉú:", X_scaled.shape)
print("PCA Î≥ÄÌôò ÌõÑ Îç∞Ïù¥ÌÑ∞ ÌòïÌÉú:", X_pca.shape)

# ÏÑ§Î™ÖÎêú Î∂ÑÏÇ∞ ÎπÑÏú® Ï∂úÎ†•: Îëê Ï£ºÏÑ±Î∂ÑÏù¥ ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÏÇ∞ÏùÑ ÏñºÎßàÎÇò ÏÑ§Î™ÖÌïòÎäîÏßÄ ÌôïÏù∏
print("Ï£ºÏÑ±Î∂ÑÎ≥Ñ ÏÑ§Î™ÖÎêú Î∂ÑÏÇ∞ ÎπÑÏú®:", pca.explained_variance_ratio_)
print("Ï¥ù ÏÑ§Î™ÖÎêú Î∂ÑÏÇ∞ ÎπÑÏú®:", np.sum(pca.explained_variance_ratio_))


# 4. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', alpha=0.8, edgecolors='w')

plt.title('4D Function Visualization using PCA', fontsize=16)
plt.xlabel('Principal Component 1', fontsize=12)
plt.ylabel('Principal Component 2', fontsize=12)
plt.grid(True)

# Î≤îÎ°Ä(Legend) Ï∂îÍ∞Ä
legend1 = plt.legend(*scatter.legend_elements(), title="Groups (y value)")
plt.gca().add_artist(legend1)

plt.show()
```
