# Level Set

### iso-contour (등위곡선) : {(x,y)|f(x,y) = h; h는 상수. 아래도 마찬가지}
### iso-surface (등위곡면) : {(x,y)|f(x,y,z) = h}
### level set : {f=h} = {x in D | f(x) = h; 여기서 x는 단변수가 아니라 벡터 in R^n}

```python
#다변수 함수 그리기

#1. 곡면 그대로 그리는 법

## f(x,y) = sin(sqrt(x^2+y^2))의 그래프 그리기
import numpy as np
import  matplotlib.pyplot as plt

x = np.linspace(-5, 5)
y = np.linspace(-5, 5)
xx, yy = np.meshgrid(x, y)
zz = np.sin(np.sqrt(xx**2+yy**2))

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(xx, yy, zz, cmap='jet')
plt.show()
```

```python
# np.meshgrid(x, y)는 두 개의 1차원 배열 x와 y를 받아서, 2차원 좌표 평면을 구성하는 격자(grid)를 만들어주는 함수.
# 각각 (x,y) 곱집합의 배열에서 x좌표(xx)와 y좌표(yy)만 나타낸 것

x = np.array([1, 2, 3])
y = np.array([10, 20])
xx, yy = np.meshgrid(x, y)

print(xx) #이차원 배열의 x좌표만 모은 것
print()
print(yy) #이차원 배열의 y좌표만 모은 것
print()
```

🔍 왜 이런 구조가 되는가?

x는 3개 → x축에 3개의 점

y는 2개 → y축에 2개의 점

그래서 전체 좌표는 총 2행 × 3열 = 6개 점이 생겨야 해.

y\x	  1	       2	  3
10	(1,10)	(2,10)	(3,10)
20	(1,20)	(2,20)	(3,20)
이걸 표현하려면:

xx: 각 위치의 x값 → 각 행이 x의 복사본

yy: 각 위치의 y값 → 각 열이 y의 복사본

즉, xx[i][j]와 yy[i][j]는 (x_j, y_i)를 나타내는 거야.

```python
#다변수 함수 그리기

#2. Level Set을 활용하는 법

import numpy as np
import matplotlib.pyplot as plt

# 위와 동일하게 이변수 함수 만드는 코드. 그래프로 시각화만 안 했을 뿐
x = np.linspace(-5, 5)
y = np.linspace(-5, 5)
xx, yy = np.meshgrid(x,y)
zz = np.sin(np.sqrt(xx**2+yy**2))

fig = plt.figure()
ax = plt.axes() # 여기서 부터 달라진다.
ax.contour(xx, yy, zz, 10, cmap='jet') #핵심차이. 4번재 인수인 10은 레벨(등고선) 개수. cmap은 색상
ax.set_aspect('equal') #가로축 세로축 비율 같게
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes() 
ax.contour(xx, yy, zz, [-0.4, -0.2, 0.0, 0.2, 0.4], cmap='jet') #4번째 인수에 리스트내 5개의 원소(f=h역할의 h 5개)로 레벨셋 부여
ax.set_aspect('equal') 
plt.show()
```

```python
#cmap: jet, bwr, brg, rainbow, gray
fig = plt.figure()
ax = plt.axes() 
ax.contour(xx, yy, zz, 10, cmap='bwr') 
ax.set_aspect('equal') 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 예시 3변수 함수: f(x, y, z) = x**2 + y**2 + z**2
def f(x, y, z):
    return x**2 + y**2 + z**2

x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
xx, yy = np.meshgrid(x, y)

# z 값을 고정하며 여러 개의 등고선 플롯 생성
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

z_values = [0, 2, 4] # 확인할 z 값들

for i, z_val in enumerate(z_values):
    zz = f(xx, yy, z_val)
    ax = axes[i]
    contour = ax.contour(xx, yy, zz, cmap='jet')
    ax.set_title(f'Contour plot at z = {z_val}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_aspect('equal')
    fig.colorbar(contour, ax=ax)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs # 가상 데이터 생성을 위한 라이브러리

# 1. 가상 데이터 생성
# 4개의 특성(변수)을 가진 3개의 클러스터(그룹) 데이터 생성
X, y = make_blobs(n_samples=300, n_features=4, centers=3, cluster_std=2.0, random_state=42)

# X는 (300, 4) 크기의 배열 (입력 데이터)
# y는 (300,) 크기의 배열 (0, 1, 2 중 하나의 값을 가지는 그룹 정보)

# 2. 데이터 정규화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. PCA 적용 (4차원 -> 2차원)
# n_components=2 : 데이터를 2개의 주성분으로 축소
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled) # PCA를 적용하여 변환

# 주성분으로 변환된 데이터 확인
print("원본 데이터 형태:", X_scaled.shape)
print("PCA 변환 후 데이터 형태:", X_pca.shape)

# 설명된 분산 비율 출력: 두 주성분이 원본 데이터의 분산을 얼마나 설명하는지 확인
print("주성분별 설명된 분산 비율:", pca.explained_variance_ratio_)
print("총 설명된 분산 비율:", np.sum(pca.explained_variance_ratio_))


# 4. 시각화
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', alpha=0.8, edgecolors='w')

plt.title('4D Function Visualization using PCA', fontsize=16)
plt.xlabel('Principal Component 1', fontsize=12)
plt.ylabel('Principal Component 2', fontsize=12)
plt.grid(True)

# 범례(Legend) 추가
legend1 = plt.legend(*scatter.legend_elements(), title="Groups (y value)")
plt.gca().add_artist(legend1)

plt.show()
```

### PCA를 이용한 4변수 함수 시각화 예제

##### 시나리오: 4개의 독립 변수 (x1, x2, x3, x4)와 이들로부터 계산되는 종속 변수 y가 있다고 가정합니다. 우리의 목표는 4차원 공간에 있는 입력 데이터 (x1, x2, x3, x4)를 2차원 평면에 나타내고, 각 점의 y 값을 색상으로 표현하여 y가 입력 공간에서 어떻게 분포하는지 확인하는 것입니다.

1) 가상 데이터 생성: 4개의 변수 (x1, x2, x3, x4)를 가진 가상의 데이터셋을 만듭니다. 이 데이터셋에는 세 개의 다른 그룹(클래스)이 존재한다고 가정하고, y 값은 이 그룹에 따라 달라지도록 설정합니다.
2) 데이터 정규화: PCA는 각 변수의 스케일에 영향을 받으므로, 모든 변수를 평균 0, 분산 1을 갖도록 표준 정규화(Standard Scaling)를 진행합니다.
3) PCA 적용: 사이킷런(scikit-learn) 라이브러리를 사용해 4차원 입력 데이터를 정보 손실을 최소화하는 2개의 주성분(PC1, PC2)으로 축소합니다.
4) 시각화: PCA로 변환된 2차원 데이터 (PC1, PC2)를 x, y 좌표로 사용하여 산점도(scatter plot)를 그립니다. 이때, 각 점의 색깔은 원래의 y 값(또는 데이터 그룹)으로 지정하여 표현합니다.
